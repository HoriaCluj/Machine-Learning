Final Version of the **Geometric Brownian Motion** Function:

```python
def gbm ():
  S_0 = 100
  r = 0.0493
  sigma = 0.1743
  S = [S_0]
  t = np.arange(1, 251, 1)
  Z = np.random.normal(0, 1, len(t))
  dt = 1 / len(t)

  for i in range(249):
    S.append(S[-1] * np.exp((r - (sigma**2)/2)*dt + sigma*np.sqrt(dt)*Z[i]))
  return np.array(S)
```
