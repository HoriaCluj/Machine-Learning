# **Black-Scholes Option Pricer**

### **Formulas**

The Black-Scholes Model assumes the stock price `St` at time `t` is given by the Geometric Brownian Motion (Solution of the SDE).

- For the European Call: `EC(K,T)` its price is given by the explicit formula:
**`EC(K,T) = exp(-qT)S0*N(d1) - Kexp(-rT)N(d2)`**

- For the European Put: **`EP(K,T) = -exp(-qT)S0*N(-d1) + Kexp(-rT)N(-d2)`**

With:
**`d1 = (log(S0/K) + (r - q - σ^2/2)T) / σ√T`**

**`d2 = d1 - σ√T`**

**`N(x)`** is the Cumulative Distribution Function (CDF) of the Standard Normal Distribution ~ N(0,1)

Therefore, the input variables are:

**`q`** Continuously Compounded Annual Dividend Yield, **`T`** Time to Maturity, **`S0`** Current Stock Price, **`K`** Strike Price, **`σ`** Annualized Volatility of the Stock, **`r`** Continuously Compounded Annual Interest Rate

As such, the Black-Scholes Option Pricer will output the No-Arbitrage (Risk-Neutral) Price for the European Call and European Put.

To **Remember**: Under non-negative interest rates (**`r>=0`**) and a non-dividend paying stock (**`q = 0`**), the Price of a European Call = American Call. Whilst under non-positive interest rates (**`r<=0`**) and no dividend yield (**`q = 0`**), the Price of a European Put = American Put.

```python
q = 0
r = 0.05
T = 1
S0 = 99.98
K = 100
sigma = 0.2
```

Implementing the **Black-Scholes Pricing Formula**:

`norm.cdf()` for the Standard Normal CDF

`math.log()` for the Natural Logarithm (ln)

```python
import numpy as np
import math
from scipy.stats import norm
```

```python
d1 = (math.log(S0/K) + (r - q + (sigma**2)/2)*T)/(sigma*np.sqrt(T))
d2 = d1 - sigma*np.sqrt(T)
```

```python
EC = np.exp(-q*T)*S0*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)
EP = -np.exp(-q*T)*S0*norm.cdf(-d1) + K*np.exp(-r*T)*norm.cdf(-d2)

print("The No-Arbitrage Price of the European Call is", round(float(EC), 2), "$")
print("The No-Arbitrage Price of the European Put is", round(float(EP), 2), "$")
```

Implementing all of the above in a function: **Black-Scholes Pricing Model**

```python
def BlackScholes(q, r, T, S0, K, sigma):
  d1 = (math.log(S0/K) + (r - q + (sigma**2)/2)*T)/(sigma*np.sqrt(T))
  d2 = d1 - sigma*np.sqrt(T)

  EC = np.exp(-q*T)*S0*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)
  EP = -np.exp(-q*T)*S0*norm.cdf(-d1) + K*np.exp(-r*T)*norm.cdf(-d2)

  print("The No-Arbitrage Price of the European Call is", round(float(EC), 2), "$")
  print("The No-Arbitrage Price of the European Put is", round(float(EP), 2), "$")

BlackScholes(0, 0.05, 1, 99.98, 100, 0.2)
```

Nonetheless, if we were to price **American Options**, both Calls and Puts, under a non-negative risk-free rate (`r >= 0)`:
- For **American Calls**: If the dividend yield `q` is zero or very small, `AC = EC` under a positive risk-free rate, therefore the price is correctly given by the Black-Scholes Pricing Model.
- To **American Puts** we need to construct a **Binomial Tree Model** with N > 100 nodes (ideally 500), where at each node the algorithm must determine the correct option price
